```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://aopwiki.org/aops'  # 실제 URL로 교체하세요
response = requests.get(url, verify=False)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table')
rows = table.find_all('tr')

output = []

for row in rows:
    cols = row.find_all(['td', 'th'])
    col_elements = [col for col in cols]

    # <li>가 포함된 컬럼들의 인덱스와 그 안의 모든 <li> 텍스트 추출
    list_cols = []
    list_col_items = []

    for i, col in enumerate(col_elements):
        lis = col.find_all('li')
        if lis:
            list_cols.append(i)
            list_col_items.append([li.get_text(strip=True) for li in lis])

    if list_cols:
        # 모든 <li> 컬럼의 항목 개수가 다를 수 있으므로 가장 많은 항목 기준으로 행 분리
        max_len = max(len(items) for items in list_col_items)

        for idx in range(max_len):
            new_row = []
            for col_idx, col in enumerate(col_elements):
                if col_idx in list_cols:
                    # 해당 <li> 컬럼의 항목 가져오기, 없으면 빈 문자열
                    li_items = list_col_items[list_cols.index(col_idx)]
                    val = li_items[idx] if idx < len(li_items) else ''
                    new_row.append(val)
                else:
                    new_row.append(col.get_text(strip=True))
            output.append(new_row)

    else:
        # <li> 없는 일반 행
        output.append([col.get_text(strip=True) for col in col_elements])

# CSV 저장
with open('aops.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(output)
```
